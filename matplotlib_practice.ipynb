```python
%matplotlib inline
```

```python
# Import Dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## Line Plot

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, label="Boston", color="green", marker="*", linewidth=2)

ax.set_xlabel("Date")
ax.set_ylabel('Fares ($)')
ax.set_ylim(0, 45)

ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.legend()
ax.grid()
```

## Bar Chart

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

## Scatter Plot

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c="red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare ($)")
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
```

## Bubble Chart

### By changing the size of the marker

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []

for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

### List Comprehension

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

## Scatter Plot using Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

## Bubble Chart using Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="skyblue", alpha=0.8, edgecolor="black", linewidth=2, label="Boston", s=[i*5 for i in y_axis])
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.set_xlim(0, 50)
ax.invert_yaxis()
ax.legend()
```

## Pie Chart

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", 
          "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Explore a pie wedge, pop out month of July.
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)


# Adjust the chart size.
plt.subplots(figsize=(8,8))

# Show the Chart.
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

## Pie Chart Using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax(figsize=(8,8))
ax.pie(y_axis,
       labels=x_axis, 
       autopct='%.1f%%')
plt.show()
```
